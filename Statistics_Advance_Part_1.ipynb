1. What is a random variable in probability theory?

3.	What is the difference between discrete and continuous distributions?

  Ans. The main difference between discrete and continuous distributions lies in the type of random variable they represent:

  ##### Discrete Distribution:

  - Represents a discrete random variable (which takes on countable values).
  - The probability of each individual value can be calculated and is non-zero.

  Examples:

  - Rolling a die: Probability of each value (e.g., P(1) = 1/6).
  - Number of heads when flipping a coin 3 times.

  ##### Continuous Distribution:

  - Represents a continuous random variable (which takes on an infinite range of values).
  - Probabilities are calculated over intervals, not individual values (since P(individual value) = 0).

  Examples:

  - Heights of people.
  - Time it takes to complete a task.

4.	What are probability distribution functions (PDF)?

  Ans. A Probability Distribution Function (PDF) describes how the probability of a random variable is distributed over its possible values. It gives insights into the likelihood of different outcomes for a random variable.

5.	How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

  Ans. The key difference between a Cumulative Distribution Function (CDF) and a Probability Distribution Function (PDF) lies in what they represent:

  ##### PDF (Probability Distribution Function):
  
  - The PDF shows the likelihood of a random variable taking on a specific value (for discrete variables) or the density of probabilities for continuous variables.

  - For example, if you're analyzing the height of people, the PDF might show that a specific range of heights is more common than others.

  - It focuses on individual values or small intervals.

  ##### CDF (Cumulative Distribution Function):

  - The CDF provides the cumulative probability that a random variable is less than or equal to a certain value.

  - It's calculated by summing up (or integrating) the PDF up to a specific point.

  - For the same example on height, the CDF tells you the probability that someone's height is less than or equal to, say, 170 cm.

  - It focuses on ranges of values.

  Why is it important?

  - It enables statisticians to use sample data to make predictions or conclusions about an entire population.

  - The CLT is the reason why the normal distribution is so significant in statistics, as it applies to many real-world scenarios.

  - It simplifies complex analyses, allowing the use of standard statistical methods for a wide variety of problems.

16.	How do you calculate a Z-score, and what does it represent?

  Ans. To calculate a Z-score, we standardize a value by measuring how far it is from the mean, expressed in terms of standard deviations. In simple terms, we compare the value to the average and account for the variability of the data.

  What Does It Represent?

  - A Z-score represents the position of a value within a distribution, showing how many standard deviations it is above or below the mean.

17.	What are point estimates and interval estimates in statistics?

  Ans. A point estimate is a precise single guess, while an interval estimate offers a broader range that reflects the reliability of the estimation.
  
  Point Estimate:
  
  - A point estimate provides a single value as the best guess for a population parameter (e.g., mean, proportion).
  - It is quick and straightforward but does not indicate the level of uncertainty in the estimate.

  Interval Estimate:

  - An interval estimate gives a range of values within which the population parameter is expected to fall, along with a level of confidence (e.g., 95% confidence interval).
  - It accounts for uncertainty and is more informative than a point estimate.

18.	What is the significance of confidence intervals in statistical analysis?

  Ans. Confidence intervals are crucial in statistical analysis because they provide a range of values within which a population parameter (like a mean or proportion) is likely to lie, along with an associated level of confidence (e.g., 95%).

  Significance:

  - Measures Uncertainty: Instead of giving a single estimate, confidence intervals quantify the uncertainty and reliability of the estimate.
  - Informs Decision-Making: They help assess whether a parameter falls within a critical range, guiding decisions in fields like business, medicine, and science.
  - Contextual Insights: Confidence intervals reveal how precise the data is and whether differences or effects are statistically significant.
  - Generalizes Results: They allow researchers to extend findings from a sample to a larger population while accounting for variability.

19.	What is the relationship between a Z-score and a confidence interval?

  Ans. The relationship between a Z-score and a confidence interval lies in how the Z-score is used to calculate the interval:

  - Z-Score Determines Confidence Level: The Z-score represents the number of standard deviations a value is from the mean. For confidence intervals, specific Z-scores correspond to desired confidence levels (e.g., a Z-score of 1.96 is used for a 95% confidence level).

  - Defines the Range: The Z-score helps determine the range (or boundaries) of the confidence interval, ensuring it captures the true population parameter within the specified level of confidence.

  - Precision and Width: Larger Z-scores (e.g., for a 99% confidence level) produce wider confidence intervals, reflecting greater certainty, while smaller Z-scores (e.g., for 90%) result in narrower intervals with less certainty.

  Z-scores are critical to constructing confidence intervals, linking the desired confidence level to the range of possible values for a population parameter.

20.	How are Z-scores used to compare different distributions?

  - Relative Position: They show where a value lies relative to its own distribution, making it easy to determine if a value is unusually high or low compared to others.

```python
def bernoulli_pdf(p, x):

    if x not in [0, 1]:
        raise ValueError("The random variable x must be either 0 or 1.")
    if not (0 <= p <= 1):
        raise ValueError("The probability p must be between 0 and 1.")

    return p if x == 1 else (1 - p)

# usage
p_success = 0.7
x_value = 1
pdf_value = bernoulli_pdf(p_success, x_value)
print(f"The PDF value for x={x_value} is: {pdf_value}")
```

18.	Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

```python
from scipy.stats import norm

def calculate_z_score(data_point, mean, std_dev):

    return (data_point - mean) / std_dev

def compare_to_standard_normal(z_score):

    cumulative_probability = norm.cdf(z_score)
    return cumulative_probability

# Example
data_point = 55
mean = 50
std_dev = 10

z_score = calculate_z_score(data_point, mean, std_dev)
print(f"The Z-score for the data point {data_point} is: {z_score:.2f}")

cumulative_prob = compare_to_standard_normal(z_score)
print(f"The cumulative probability corresponding to the Z-score is: {cumulative_prob:.4f}")
```

19.	Implement hypothesis testing using Z-statistics for a sample dataset.

```python
def hypothesis_testing(sample, population_mean, population_std, alpha=0.05):

    sample_mean = np.mean(sample)
    sample_size = len(sample)

    standard_error = population_std / np.sqrt(sample_size)
    z_statistic = (sample_mean - population_mean) / standard_error

    z_critical = norm.ppf(1 - alpha / 2)

    if abs(z_statistic) > z_critical:
        return f"Reject the null hypothesis (Z-statistic: {z_statistic:.2f})"
    else:
        return f"Fail to reject the null hypothesis (Z-statistic: {z_statistic:.2f})"

# Example dataset
sample_data = [102, 108, 95, 99, 105, 101, 94, 97, 103, 98]

result = hypothesis_testing(sample_data, population_mean=100, population_std=15, alpha=0.05)
print(result)
```

20.	Create a confidence interval for a dataset using Python and interpret the result.

```python
def calculate_confidence_interval(data, confidence=0.95):

    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)
    sample_size = len(data)
    standard_error = std_dev / np.sqrt(sample_size)

    z_score = norm.ppf(1 - (1 - confidence) / 2)

    margin_of_error = z_score * standard_error

    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error
    return lower_bound, upper_bound

# Example
data = [18, 21, 19, 22, 17, 20, 23, 19, 18, 20]

confidence_interval = calculate_confidence_interval(data, confidence=0.95)
print(f"95% Confidence Interval: {confidence_interval}")
```

28.	Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

```python
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(data):

    skewness = skew(data)
    kurt = kurtosis(data)
    return skewness, kurt

mean = 50
std_dev = 10
size = 1000

data = np.random.normal(mean, std_dev, size)

skewness, kurt = calculate_skewness_kurtosis(data)

print(f"Skewness of the dataset: {skewness:.4f}")
print(f"Kurtosis of the dataset: {kurt:.4f}")
```
